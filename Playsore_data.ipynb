```python
def convert_size(size):
    if 'k' in size:
        return float(size.replace('k', ''))
    elif 'M' in size:
        return int(float(size.replace('M', ''))) * 1000
    elif 'Varies with device' in size:
        return np.nan
    else:
        return np.nan
```

```python
df['Size'] = df['Size'].apply(convert_size)
```

```python
df.head()
```

```python
cols = ['Installs', 'Price']
chars_to_remove=['+',',','$']
for i in cols:
    for j in chars_to_remove:
        df[i] = df[i].str.replace(j, '')
```

```python
df.head()
```

```python
df['Installs'] = df['Installs'].astype('int')
df["Price"] = df['Price'].astype('float')
```

```python
df.info()
```

```python
df["Last Updated"] = pd.to_datetime(df['Last Updated'])
df['day'] = df['Last Updated'].dt.day
df['month'] = df['Last Updated'].dt.month
df['year'] = df['Last Updated'].dt.year
```

# Q11. How does the rating of apps vary by category? Create a boxplot to compare the ratings of different app categories.

```python
plt.figure(figsize=(20,10))
sns.boxplot(x=df['Category'], y=df['Rating'])
plt.xticks(rotation=90)
plt.show()
```

# Q12. What is the relationship between the size of an app and its rating? Create a scatter plot to visualize the relationship.

```python
plt.figure(figsize=(20,10))
sns.scatterplot(x='Size', y='Rating', data=df)
plt.title('App Size vs. Rating')
plt.xlabel('App Size')
plt.ylabel('Rating')
plt.xticks(rotation=90)
plt.show()
```

# Q13. How does the type of app affect its price? Create a bar chart to compare average prices by app type.

```python
plt.figure(figsize=(12, 6))
sns.barplot(x='Type', y='Price', data=df, estimator=np.mean)
plt.title('Average Price by App Type')
plt.xlabel('App Type')
plt.ylabel('Average Price')
plt.show()
```

```python
df['Type'].value_counts()
```

```python
plt.figure(figsize=(16, 8))
sns.barplot(x='Category', y='Price', data=df)
plt.xticks(rotation=90)
plt.title('Average Prices by App Category')
plt.xlabel('App Category')
plt.ylabel('Average Price')
plt.show()
```

# Q14. What are the top 10 most popular apps in the dataset? Create a frequency table to identify the apps with the highest number of installs.

```python
df.sorted = df.sort_values(by='Installs', ascending=False)
top_10_apps = df.sorted.head(10)

plt.figure(figsize=(10, 6))
plt.bar(top_10_apps['App'], top_10_apps['Installs'])
plt.xlabel('App')
plt.ylabel('Number of Installs')
plt.title('Top 10 Apps by Installs')
plt.xticks(rotation=45, ha='right')  # Rotate x-axis labels for better readability

plt.show()
df_top_apps = df.sorted 
sns.barplot(x=df_top_apps['App'], y = df_top_apps['Installs'])
```

# Q16. A mobile app development company wants to analyze the Google Playstore dataset to identify the most successful app developers. What features would you analyze to make recommendations to the company, and what data visualizations would you use to present your findings?

Ans: For this we can check with the app version and the installs. we can show it in barplot or in App Updates Over Time by line chart showing the number of app updates over time.

```python
developer_metrics = df.groupby('Android Ver').agg({'Rating': 'mean', 'Installs': 'sum'}).sort_values(by='Installs', ascending=False)
plt.figure(figsize=(10, 6))
developer_metrics.head(15).plot(kind='bar', y='Installs', color='blue', 
                                title='Top 15 Android Version by Total Installs', 
                                xlabel='Version', 
                                ylabel='Total Installs',
                                rot=90)
```
